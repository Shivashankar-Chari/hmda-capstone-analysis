```python
import pandas as pd

file_path = "2024_combined_mlar.zip"

df = pd.read_csv(
    file_path,
    sep="|",
    compression="zip",
    header=None,   
    low_memory=False
)

print(df.shape)
df.head()
```

```python
column_names = [
    "activity_year",
    "lei",
    "loan_type",
    "loan_purpose",
    "preapproval",
    "construction_method",
    "occupancy_type",
    "loan_amount",
    "action_taken",
    "state_code",
    "county_code",
    "tract",
    "applicant_ethnicity_1",
    "applicant_race_1",
    "applicant_sex",
    "income",
    "purchaser_type",
    "rate_spread",
    "hoepa_status",
    "lien_status"
]

df.columns = column_names + [f"extra_{i}" for i in range(len(df.columns) - len(column_names))]
```

```python
import matplotlib.pyplot as plt

df['action_taken'].value_counts().plot(kind='bar')
plt.title("Loan Decision Distribution")
plt.show()
```

```python
import pandas as pd

file_path = "2024_combined_mlar_header.zip"

filtered_chunks = []

for chunk in pd.read_csv(
    file_path,
    sep="|",
    compression="zip",
    chunksize=200000,
    low_memory=False
):
    # Keep only relevant outcomes
    filtered = chunk[chunk['action_taken'].isin([1, 3])]
    filtered_chunks.append(filtered)

# Combine all filtered chunks
df_full = pd.concat(filtered_chunks, ignore_index=True)
```

```python
print(df_full.shape)
df_full.head()
```

```python
df_full = df_full.sample(n=300000, random_state=42)
```

```python
import pandas as pd

file_path = "2024_combined_mlar_header.zip"

use_cols = [
    'loan_amount',
    'income',
    'action_taken',
    'applicant_race_1',
    'applicant_sex',
    'debt_to_income_ratio',
    'state_code',
    'interest_rate'
]

filtered_chunks = []

for chunk in pd.read_csv(
    file_path,
    sep="|",
    compression="zip",
    chunksize=200000,
    usecols=use_cols,
    low_memory=False
):
    # Filter only approved (1) and denied (3)
    filtered = chunk[chunk['action_taken'].isin([1, 3])]
    
    filtered_chunks.append(filtered)

# Combine
df_small = pd.concat(filtered_chunks, ignore_index=True)

# Sample
df_small = df_small.sample(n=300000, random_state=42)

print(df_small.shape)
df_small.head()
```
